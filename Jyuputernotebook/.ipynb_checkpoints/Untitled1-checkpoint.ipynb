```python
from bs4 import BeautifulSoup
import requests

a  = requests.get("https://twitter.com/shigure_bsk")
b = BeautifulSoup(a.text,'html.parser')
c = b.find_all('span','href')

for soup in c:
    print(soup.getText())
```
