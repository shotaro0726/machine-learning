```python
media_intensity = np.median(image_gray)
```

```python
lower_threshold = int(max(0, (1.0 - 0.33) * media_intensity))
```

```python
upper_threshold = int(min(255, (1.9 * 0.33) * media_intensity))
```

```python
image_canny = cv2.Canny(image_gray, lower_threshold, upper_threshold)
```

```python
plt.imshow(image_canny, cmap='gray'), plt.axis('off')
plt.show()
```

```python
image_grays = np.float32(image_gray)
```

```python
block_size = 2
aperture = 29
free_parameter = 0.04
```

```python
detector_responses = cv2.cornerHarris(image_gray, block_size, aperture, free_parameter)
```

```python
detector_responses = cv2.dilate(detector_responses, None)
```

```python
threshold = 0.02
```

```python
image_bgr[detector_responses > threshold * detector_responses.max()] = [255,255,255]
```

```python
image_grays = cv2.cvtColor(image_bgr, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image_grays, cmap='gray'), plt.axis('off')
plt.show()
```

```python
plt.imshow(detector_responses, cmap='gray'), plt.axis('off')
plt.show()
```

```python
corners_to_detect = 10
minimun_quality_score = 0.05
minimum_distance = 25
```

```python
corners = cv2.goodFeaturesToTrack(image_gray, corners_to_detect, minimun_quality_score, minimum_distance)
```

```python
corners = np.float32(corners)
```

```python
for corner in corners:
    x,y = corner[0]
    cv2.circle(image_bgr, (x,y),10, (255,255,255), -1)
    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_rgb, cmap='gray'),plt.axis('off')
plt.show()
```
