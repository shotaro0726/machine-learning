```python
import csv
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.nukistream.com/"
html = requests.get(url)
soup = BeautifulSoup(html.text,'html.parser')
a_tag = soup.find_all("a")

csvFile = open("nuki.csv",'wt',encoding='utf-8')
writer = csv.writer(csvFile)

for a in a_tag:
    csvRow = []
    csvRow.append(a.getText())
    writer.writerow(csvRow)
    print(csvRow)

csvFile.close()
```
